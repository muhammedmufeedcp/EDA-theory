Standardization and normalization are two ways to rescale data.

Standardization rescales a dataset to have a mean of 0 and a standard deviation of 1.

It uses the following formula to do so:

Xnew = (X -Meu)/Sigma,   X = the each data value, Meu = Mean of each feature
Sigma = Standard Deviation

Normalization rescales a dataset so that each value falls between 0 and 1. It uses the following formula

Xnew = (X - Xmin)/(Xmax - Xmin)     , X = the each data value, Xmin = Minimum

value of a column, Xmax = Maximum value of a column
